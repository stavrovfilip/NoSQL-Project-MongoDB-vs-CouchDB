```python

listJsonColumns = ['genres','production_companies','production_countries','spoken_languages']

for col in listJsonColumns:
  for i in range(0,len(data[col])):
    if pd.isnull(data[col][i]) == False:
      data[col][i] = eval(data[col][i])
  
```
